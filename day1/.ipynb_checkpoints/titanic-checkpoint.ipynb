```python
data.loc[ (data['Age'].isnull()) & (data['Initial'] == 'Mr'), 'Age' ] = 33
data.loc[ (data['Age'].isnull()) & (data['Initial'] == 'Master'), 'Age' ] = 5
data.loc[ (data['Age'].isnull()) & (data['Initial'] == 'Mrs'), 'Age' ] = 36
data.loc[ (data['Age'].isnull()) & (data['Initial'] == 'Miss'), 'Age' ] = 22
data.loc[ (data['Age'].isnull()) & (data['Initial'] == 'Other'), 'Age' ] = 46
```

```python
data.isna().sum()
```

```python
def get_category(age):
    cat = ''
    if age <= -1 : cat='Unknown'
    elif age <= 5 : cat = 'Baby'
    elif age <= 12 : cat = 'Child'
    elif age <= 18 : cat = 'Teenager'
    elif age <= 35 : cat = 'AAAAA'
    elif age <= 60 : cat = 'BBBBBB'
    else : cat = 'Elderly'
        
    return cat
```

```python
data['Age_cat'] = data['Age'].apply(lambda x : get_category(x))
```

```python
data['Age_cat'].value_counts()
```

```python
data.Embarked.value_counts()
```

```python
mapping = {
    'S' : 0,
    'C' : 1,
    'Q' : 2
}

data['Embarked'] = data['Embarked'].map(mapping)
```

```python
data.Embarked.value_counts().plot(kind='bar')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def encode_features(data_df):
    features = ['Sex', 'Age_cat']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(data_df[feature])
        data_df[feature] = le.transform(data_df[feature])
    return data_df
```

```python
data.head(1)
```

```python
data = encode_features(data)
data.head()
```
