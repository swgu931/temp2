# 시리즈, 데이터프레임 

시리즈 : 컬럼이 하나.  
데이터프레임 : 컬럼이 여러개

```python
col  = ['col1']
array = np.array([1,2,3])
```

```python
d = pd.DataFrame(array, columns = col)
d
```

```python
# 2차원 numpy -> dataframe
col = ['col1', 'col2', 'col3']
array = np.array([[1,2,3],
                 [4,5,6]])

d = pd.DataFrame(array, columns = col)
d
```

```python
#딕셔너리 -> dataframe
d = {
    'col1' : [1,2],
    'col2' : [3,4],
    'col3' : [5,6]
}
d = pd.DataFrame(d)
d
```
