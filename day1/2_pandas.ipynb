**정렬하기**

**Aggregation 함수 적용**

min(), max(), sum(), count()와 같은 aggregation 함수 적용은 RDBMS SQL의 aggregation 함수 적용과 유사하다.  
하지만 dataframe의 경우 바로 aggregation을 호출할 경우 모든 칼럼에 agg를 적용한다는 특징이 있다.  
아래와 같이.

```python
data_copy.count()
```

특정 컬럼에 적용하려면 아래와 같이 진행한다.

```python
data_copy[['Age', 'SibSp']].mean()
```

**groupby() 적용**

dataframe의 groupby는 RDBMS SQL의 groupby와 유사하면서도 다르다.  
groupby 사용 시 입력 파라미터로 by를 기준으로 그룹을 묶어준다.

```python
data_copy_group = data_copy.groupby(by='Pclass')
data_copy_group
```

이렇게 groupby를 하면 DataFrameGroupBy라는 객체가 반환되어서 생성된다.  
SQL과 다르게 groupby에 적용된 객체에 aggregation을 적용하면 모든 컬럼에 적용한다.

```python
data_copy_group.count()
```

만약, 이것을 SQL로 했다면 SELECT COUNT(PassengerId), COUNT(Survived).. FROM data_copy GROUP BY Pclass 등으로 해야 했을 것이다.  

특정 컬럼에 agg를 적용하고 싶으면 특정 컬럼만 빼서 하면 된다.

```python
data_copy_group['Fare'].max()
```

```python
data_copy_group[['Fare', 'Age']].count()
```

그리고 SQL에서 SELECT MAX(Age), MIN(Age) FROM data_copy GROUP BY Pclass와 같은 SQL을 pandas groupby에서는 아래와 같이 할 수 있다.

```python
data_copy.groupby(by='Pclass')['Age'].agg([min, max])
```

그러면 여러개의 컬럼에 적용은 어떻게 시킬까?  
예를 들어서 SELECT MAX(Age), SUM(Fare) FROM data_copy GROUP BY Pclass와 같은 SQL을 만드려면?

```python
agg_d = {
    'Age' : 'max',
    'Fare' : 'count'
}

data_copy.groupby(by='Pclass').agg(agg_d)
```

**별도 함수 적용**

```python
def get_category(age):
    c = ''
    if age <= 5 : c = 'baby'
    elif age <= 12 : c = 'child'
    elif age <= 18 : c = 'teenager'
    elif age <= 25 : c = 'student'
    elif age <= 35 : c = 'young adult'
    elif age <= 60 : c = 'adult'
    else : c = 'elderly'
    return c
```

```python
data_copy['Age_c3'] = data_copy['Age'].apply(lambda x : get_category(x))
data_copy.head()
```
